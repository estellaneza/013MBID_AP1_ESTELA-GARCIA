#### AP1 - Limpieza de los datos

---

En esta libreta se realiza una evaluación básica de calidad de los datos del escenario y se ejecutan acciones de limpieza.

```python
# Se importan las librerías necesarias y se suprimen las advertencias
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore',category=FutureWarning)
warnings.filterwarnings('ignore',category=UserWarning)
```

Lectura del mismo dataset inicial

```python
# Lectura de los datos
df_creditos = pd.read_csv("../data/raw/datos_creditos.csv", sep=";")
display(df_creditos.head(5))

df_tarjetas = pd.read_csv("../data/raw/datos_tarjetas.csv", sep=";")
display(df_tarjetas.head(5))
```

##### Transformaciones sobre los datos

Verificación de nulos

```python
# Se verifica la cantidad de valores nulos en cada columna
print("Valores nulos en el dataset de créditos:")
for col in df_creditos.columns:
   print(f"Atributo: {col} - Valores nulos: {df_creditos[col].isna().sum()} ({df_creditos[col].isna().sum()/len(df_creditos)*100:.2f}%)")
```

```python
# Se verifica la cantidad de valores nulos en cada columna
print("Valores nulos en el dataset de tarjetas:")
for col in df_tarjetas.columns:
   print(f"Atributo: {col} - Valores nulos: {df_tarjetas[col].isna().sum()} ({df_tarjetas[col].isna().sum()/len(df_tarjetas)*100:.2f}%)")
```

Selección de datos (filtro de columnas)

```python
# Se establece qué columnas se eliminan
col_eliminar_creditos = []
col_eliminar_tarjetas = ['nivel_tarjeta']

# Se ejecuta la operación
df_creditos.drop(col_eliminar_creditos, inplace=True, axis=1)
df_tarjetas.drop(col_eliminar_tarjetas, inplace=True, axis=1)
```

Limpieza de datos (filtro de filas)

```python
# -------------------------------------------------
# Limpieza de datos (filtro de filas)
# -------------------------------------------------

# Copia del dataset original
df_creditos_filtrado = df_creditos.copy()

# -------------------------------------------------
# COMPROBACIÓN DE NULOS ANTES DEL TRATAMIENTO
# -------------------------------------------------
print("NULOS ANTES DE LA IMPUTACIÓN")
print("tasa_interes:", df_creditos_filtrado['tasa_interes'].isna().sum())
print("antiguedad_empleado:", df_creditos_filtrado['antiguedad_empleado'].isna().sum())

# -------------------------------------------------
# 1. Imputación de valores nulos en tasa_interes
# Mediana por grupo de objetivo_credito
# -------------------------------------------------
df_creditos_filtrado['tasa_interes'] = (
    df_creditos_filtrado
    .groupby('objetivo_credito')['tasa_interes']
    .transform(lambda x: x.fillna(x.median()))
)

# -------------------------------------------------
# 2. Imputación de valores nulos en antiguedad_empleado
# Mediana por grupo de edad
# -------------------------------------------------
df_creditos_filtrado['antiguedad_empleado'] = (
    df_creditos_filtrado
    .groupby('edad')['antiguedad_empleado']
    .transform(lambda x: x.fillna(x.median()))
)
# -------------------------------------------------
# COMPROBACIÓN DE NULOS DESPUÉS DEL TRATAMIENTO
# -------------------------------------------------
print("\nNULOS DESPUÉS DE LA IMPUTACIÓN")
print("tasa_interes:", df_creditos_filtrado['tasa_interes'].isna().sum())
print("antiguedad_empleado:", df_creditos_filtrado['antiguedad_empleado'].isna().sum())

# -------------------------------------------------
# Filtrado de edades mayores de 90 años
# -------------------------------------------------
df_creditos_filtrado = df_creditos_filtrado[df_creditos_filtrado['edad'] < 90]

# -------------------------------------------------
# Eliminación final de registros que aún tengan nulos
# -------------------------------------------------
filas_antes = df_creditos_filtrado.shape[0]
df_creditos_filtrado.dropna(inplace=True)
```

Integración de ambos datasets en uno solo

```python
df_integrado = pd.merge(df_creditos_filtrado, df_tarjetas, on='id_cliente', how='inner')

print(f"Filas del dataset integrado con los filtros realizados: {df_integrado.shape[0]}")
print(f"Columnas del dataset integrado post integración: {df_integrado.shape[1]}")
print(f'El dataset original tenía {df_creditos.shape[0]} filas en Créditos y {df_tarjetas.shape[0]} filas en Tarjetas.')
print(f'Se han eliminado {df_creditos.shape[0] - df_creditos_filtrado.shape[0]} filas en total.')
```

```python
# -------------------------------------------------
# Generación de atributos derivados (impresión forzada)
# -------------------------------------------------

print("="*80)
print("INICIO GENERACIÓN DE ATRIBUTOS DERIVADOS")
print("="*80, "\n")

# =================================================
# CRÉDITOS
# =================================================

# 1. Capacidad de pago
df_integrado['capacidad_pago'] = (
    df_integrado['importe_solicitado'] / df_integrado['ingresos']
)

print("✅ Atributo generado: capacidad_pago")
print(df_integrado['capacidad_pago'].describe())
print("\n" + "-"*80 + "\n")


# 2. Estabilidad laboral
df_integrado['estabilidad_laboral'] = (
    df_integrado['antiguedad_empleado'] / df_integrado['edad']
)

print("✅ Atributo generado: estabilidad_laboral")
print(df_integrado['estabilidad_laboral'].describe())
print("\n" + "-"*80 + "\n")


# =================================================
# TARJETAS
# =================================================

# 3. Operaciones mensuales
df_integrado['operaciones_mensuales'] = (
    df_integrado['operaciones_ult_12m'] / 12
)

print("✅ Atributo generado: operaciones_mensuales")
print(df_integrado['operaciones_mensuales'].describe())
print("\n" + "-"*80 + "\n")


# 4. Gasto medio mensual
df_integrado['gasto_medio_mensual'] = (
    df_integrado['gastos_ult_12m'] / 12
)

print("✅ Atributo generado: gasto_medio_mensual")
print(df_integrado['gasto_medio_mensual'].describe())
print("\n" + "-"*80 + "\n")


# 5. Gasto promedio por operación
df_integrado['gasto_promedio_operacion'] = (
    df_integrado['gastos_ult_12m'] / df_integrado['operaciones_ult_12m']
)

print("✅ Atributo generado: gasto_promedio_operacion")
print(df_integrado['gasto_promedio_operacion'].describe())
print("\n" + "-"*80 + "\n")


# =================================================
# VERIFICACIÓN FINAL
# =================================================

print("COLUMNAS NUEVAS EN EL DATASET:")
print([
    'capacidad_pago',
    'estabilidad_laboral',
    'operaciones_mensuales',
    'gasto_medio_mensual',
    'gasto_promedio_operacion'
])

print("\nVISTA PREVIA:")
display(
    df_integrado[[
        'capacidad_pago',
        'estabilidad_laboral',
        'operaciones_mensuales',
        'gasto_medio_mensual',
        'gasto_promedio_operacion'
    ]].head(10)
)

print("\nFIN GENERACIÓN DE ATRIBUTOS")
print("="*80)
```

Transformaciones sobre atributos

```python
# Antiguedad del empleado

etiquetas_a_e = ['menor_5', '5_a_10', 'mayor_10']
rangos_a_e = [0, 4, 10, 50]
valor_para_nan = 'NA'
antiguedad_empleados_N = pd.cut(df_integrado['antiguedad_empleado'], 
                                bins=rangos_a_e, 
                                labels=etiquetas_a_e,
                                right=False).cat.add_categories(valor_para_nan).fillna(valor_para_nan)

display(antiguedad_empleados_N.value_counts())

# edad

etiquetas_e = ['menor_25', '25_a_30']
rangos_e = [0, 24, 50]
edad_N = pd.cut(df_integrado['edad'], 
                                bins=rangos_e, 
                                labels=etiquetas_e)

display(edad_N.value_counts())
```

```python
# -------------------------------------------------
# Transformaciones sobre atributos
# Eliminación de columnas redundantes o no informativas
# -------------------------------------------------

# Definición de columnas a eliminar (versión mejorada)
columnas_a_eliminar = [
    "id_cliente",            # Identificador técnico
    "operaciones_ult_12m",   # Sustituida por operaciones_mensuales
    "importe_solicitado",    # Integrada en capacidad_pago
    "duracion_credito",      # Variable auxiliar con baja aportación
    "gastos_ult_12m",        # Sustituida por gasto_medio_mensual y gasto_promedio_operacion
    "antiguedad_empleado"    # Integrada en estabilidad_laboral
]

print("Columnas antes de eliminar:")
print(df_integrado.columns.tolist(), "\n")

# Eliminación de columnas
df_integrado.drop(columns=columnas_a_eliminar, inplace=True)

print("Columnas eliminadas:")
print(columnas_a_eliminar, "\n")

print("Columnas finales del dataset:")
print(df_integrado.columns.tolist())
```

Validaciones finales

```python
# Se verifica la cantidad de valores nulos en cada columna del dataset integrado
print("Valores nulos en el dataset integrado:")
for col in df_integrado.columns:
   print(f"Atributo: {col} - Valores nulos: {df_integrado[col].isna().sum()} ({df_integrado[col].isna().sum()/len(df_integrado)*100:.2f}%)")

# Se verifica la presencia de valores duplicados en el dataset integrado
num_duplicados = df_integrado.duplicated().sum()
print(f"\nNúmero de filas duplicadas en el dataset integrado: {num_duplicados}")
```

```python
# -------------------------------------------------
# Resumen de métricas relevantes del dataset integrado
# -------------------------------------------------

num_filas = df_integrado.shape[0]

resumen = pd.DataFrame({
    "Atributo": df_integrado.columns,
    "Tipo de Dato": df_integrado.dtypes.values,
    "Valores Únicos": df_integrado.nunique().values,
    "Porcentaje Únicos (%)": (df_integrado.nunique() / num_filas * 100).round(2),
    "Valores Nulos": df_integrado.isna().sum().values,
    "Porcentaje Nulos (%)": (df_integrado.isna().sum() / num_filas * 100).round(2)
})

display(resumen)
```
